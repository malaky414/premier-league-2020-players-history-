```python
manutd = df[df["Club"] == "Manchester-United"].copy()
```

```python
manutd_shots = len(manutd)
manutd_goals = manutd["Goals"].sum()
manutd_BC = manutd["Big chances created"].sum()

manutd_shots, manutd_goals, manutd_BC
```

```python
plt.figure()
plt.bar(["Goals", "Big Chances"], [manutd_goals, manutd_BC])
plt.title("Manchester United: Goals vs Big Chances")
plt.show()
```

```python
player_summary = (
    manutd.groupby("Name")
    .agg(
        shots=("Goals", "count"),
        goals=("Goals", "sum"),
        BigChances=("Big chances created", "sum")
    )
    .reset_index()
)

player_summary["BigChances"] = player_summary["goals"] - player_summary["BigChances"]
player_summary.sort_values("BigChances").head()
```

```python
# here we wanna test shot accuracy for the best attackers whose shots less and score more
# so we choose the players which plaied more than 10 matches and scored more than 10 goals
df = df[(df["Appearances"] >= 10) & (df["Goals"]>=10)]
```

```python
df["goal_conversion"] = df["Goals"] / df["Shots"]
df["goals_per_app"] = df["Goals"] / df["Appearances"]
df["attacking_contribution"] = (df["Goals"] + df["Assists"]) / df["Appearances"]
df["big_chance_waste"] = df["Big chances missed"] / df["Appearances"]
```

```python
attackers = df[df["Position"].isin(["Forward", "Midfielder"])]
```

```python
attackers.sort_values("goal_conversion", ascending=False)[
    ["Name", "Club", "Goals", "Shots", "goal_conversion"]
].head(10)
```

```python
attackers.sort_values("big_chance_waste", ascending=False)[
    ["Name", "Club", "Big chances missed", "Appearances", "big_chance_waste"]
].head(10)
```

```python
attackers.sort_values("attacking_contribution", ascending=False)[
    ["Name", "Club", "Goals", "Assists", "Appearances", "attacking_contribution"]
].head(10)
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(attackers["Shots"], attackers["Goals"], alpha=0.6)
plt.xlabel("Shots")
plt.ylabel("Goals")
plt.title("Attacking Efficiency: Goals vs Shots")
plt.show()
```
